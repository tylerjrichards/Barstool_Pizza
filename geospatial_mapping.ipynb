```python
import numpy as np 
import pandas as pd
import plotly
import plotly.plotly as py
import plotly.offline as offline
import plotly.graph_objs as go
```

```python
pizza_df = pd.read_csv("pizza_data.csv")
pizza_df['string_concat'] = pizza_df['name'] + " | Dave - " + pizza_df['reviewStats.dave.averageScore'].astype(str) + " | Community - " + round(pizza_df['reviewStats.community.averageScore'], 2).astype(str)
```

```python
pizza_df['string_concat']
```

```python
pizza_df['dave_score_int'] = [int(x*10) for x in pizza_df['reviewStats.dave.averageScore']]
```

```python
pizza_df.to_csv("pizza_data.csv", index=None)
```

```python
pizza_df['reviewStats.dave.averageScore']*10
```

```python
shaz13_custom_style = "mapbox://styles/shaz13/cjiog1iqa1vkd2soeu5eocy4i"
#set the geo=spatial data
data = [go.Scattermapbox(
            lat= pizza_df['latitude'] ,
            lon= pizza_df['longitude'],
            mode='markers',
            marker=dict(
                size= 8,
                color = 'blue',
                opacity = .8,
            ),
            text=pizza_df['string_concat'],
          )]
#set the layout to plot
layout = go.Layout(autosize=False,
                   mapbox= dict(accesstoken="",
                                bearing=10,
                                pitch=60,
                                zoom=13,
                                center= dict(lat=40.721319,
                                             lon=-73.987130),
                                style=shaz13_custom_style),
                    width=900,
                    height=600, 
                    title = "Pizza Locations from One Bite Pizza")
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

init_notebook_mode(connected=True)
fig = dict(data=data, layout=layout)
iplot(fig)
```
